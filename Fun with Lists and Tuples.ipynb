```python
L = [(1,2,5), (3,9), (4,8), (6,2), (5,4), (5,5)]
```

```python
i = 0
for i in range(i,len(L)):
    L[i] = list(L[i])
    for j in range (i+1,len(L)):        
        if L[i][-1] > L[j][-1]:
            temp = L[i]
            L[i] = L[j]
            L[j] = temp
    L[i] = tuple(L[i])             
print("Sorted List in increasing order by the last element ", L)


```
